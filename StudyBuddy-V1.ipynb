```python
# ðŸ“˜ StudyBuddy: Ask Your PDF Anything â€” Gemini API Project

# âœ… STEP 1: Install Required Libraries
!pip install -q google-generativeai PyMuPDF

# âœ… STEP 2: Import Libraries
import google.generativeai as genai
import fitz  # PyMuPDF
import os
import textwrap
from google.colab import files
from google.colab import userdata

# âœ… STEP 3: Configure Gemini API
userdata.get('GEMINI_API_KEY')
# âœ… STEP 3: Configure Gemini API
genai.configure(api_key=GEMINI_API_KEY)

# âœ… STEP 4: Upload a PDF
print("ðŸ“‚ Upload your PDF file (e.g., study notes):")
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# âœ… STEP 5: Extract text from PDF using PyMuPDF
def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

document_text = extract_text_from_pdf(pdf_path)

#for m in genai.list_models():
#   print(m.name, m.supported_generation_methods)


# âœ… STEP 6: Load Gemini Model
model = genai.GenerativeModel("gemini-2.5-pro")

# âœ… STEP 7: Ask Questions Based on PDF
def ask_gemini_about_pdf(question, context):
    prompt = f"""You are StudyBuddy, an academic assistant.
Use the context below to answer the question as accurately as possible.

Context:
\"\"\"
{context}
\"\"\"

Question: {question}
Answer:"""
    response = model.generate_content(prompt)
    return response.text.strip()

# âœ… STEP 8: Loop for Interaction
while True:
    question = input("ðŸ’¬ Ask a question about the PDF (or type 'exit'): ")
    if question.lower() == 'exit':
        break
    answer = ask_gemini_about_pdf(question, document_text)
    print("\nðŸ¤– Gemini Answer:\n")
    print(textwrap.fill(answer, width=100))
    print("-" * 80)
```
